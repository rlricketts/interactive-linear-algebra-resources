## 1. Reduced Row-Echelon Form

<p>In the previous notebook we say how we can convert a linear system into an augmented matrix and solved it.</p>

$$
\begin{vmatrix} 
    x + 2y + 3z = 39 \\
    x + 3y + 2z = 34 \\
    3x + 2y + z = 26
\end{vmatrix}
$$

$$
\left[
\begin{array}{ccc|c}
    1 & 2 & 3 & 39 \\
    1 & 3 & 2 & 34 \\
    3 & 2 & 1 & 26
\end{array}
\right]
$$

$$
\left[
\begin{array}{ccc|c}
    1 & 0 & 0 & 2.75 \\
    0 & 1 & 0 & 4.25 \\
    0 & 0 & 1 & 9.25
\end{array}
\right]
$$




<p>Formally this final form is refered to as reduced row-echelon form.  For a matrix $A$ the notation for this form is customerally as $rref(A)$
    
<p>This method of solving linear systems may be referred to as <em>Gauss-Jordan elimination</em>, after Carl Fridrich Gauss (1777-1885) and Wilhelm Jordan (1844-1899), a German mathemetician and engineer respectively. Though the technique is credited to these two there is evidence for this method being developed 2,000 years prior in China.</p>

<div class="alert alert-block alert-info">
    <strong>Reduced row-echelon form:</strong>
    <p>For a matrix to be in reduced row-eschlon form must satify <strong><u>all</u></strong> of the folowing conditions.</p>
<ol>
    <li>If a row does not consist entire of zeros, then thef rist nonzero nome n the row is a 1 (referred to as the <em>leading</em> 1 or <em>pivot</em> in this row)</li>
    <li>If a column contains a leading 1, then all the other entries in taht column are 0</li>
    <li>If a row contains a leading 1, then each row aboe it contains a leading 1 further to the left.</li>
    <li>If there are any rows that consist entirely of zeros, then they are grouped together at the bottom of the matrix</li>
</ol>
</div>

<div class="alert alert-block alert-warning">
    <p>This <em>Gauss-Jordan elimination</em> method of solving linear systems is a form of algorithm.  David Berlinski defines an algorithm as the following in <em>The Advent of the Algorithm: The Idea that Rules the World</em>:</p>

The TI calculator or software (Matlab, Octave, Python, etc.) you are using to solve linear systems do not use this exact algoritm due to the floating-point arithmatic rounding errors discussed in the previous notebook.</p>

## 2. Rank

<p>Reduced row-echelon form gives us information about another attribute of matrices, rank.</p>
<div class="alert alert-block alert-info">
    <strong>Definition of Rank:</strong>
<p>The rank of a matrix $A$ is the number of leading 1's in $rref(A)$.</p>
</div>
<p>We refer to the rank of a matrix rather than a linear system as we must specify whether the coefficent or augmented matrics of the system is under consideration.</p>
    

$$ \text{If }A =
\begin{bmatrix}
    1 & 2 & 3\\
    4 & 5 & 6\\
    7 & 8 & 9
\end{bmatrix} 
\text{, }
rref(A) =
\begin{bmatrix}
    1 & 0 & -1\\
    0 & 1 & 2\\
    0 & 0 & 0
\end{bmatrix} 
\text{, therefore }
rank(A)=2
$$

<p>While the concept of rank seems quite straightfoward, the signifcance of the relationship between the size of a matrix and its rank is signfificant.</p>

<p>We we consider the coeffeicent matrix $A$ with size $n \times m$ we can prove the following:</p>
    
<ul>
    <li>$rank(A) \leq n$ and $rank(A) \leq m$ are always true.</li>
    <li>If $rank(A) = n$, the system is consistent.</li>
    <li>If $rank(A) = m$, the sytem hs at most one solution.</li>
    <li>If $rank(A) < m$, the system has either infinitely many solutions or none</li>
    <li>(# of free variables) = (total # of variables) - (# of leading variables) = $m - rank(A)$</li>
</ul>

## 2. Problems with No Solution


<p>Given the following matrix how do we interpret its solution?</p>

$$
\begin{align}
\text{linear system} &=
\begin{vmatrix} 
    2x + 4y = 3 \\
    3x + 6y = 2 
\end{vmatrix}
\\ 
\\
A &= 
\left[
\begin{array}{cc|c}
    2 & 4 & 3 \\
    3 & 6 & 2 
\end{array}
\right]
\\ 
\\
rref(A) &= 
\left[
\begin{array}{cc|c}
    1 & 2 & 0 \\
    0 & 0 & 1
\end{array}
\right]
\end{align}
$$

<p>Initially, we notice that in regards to the coefficient matrix $rank(A) < m$, $(1<2)$, indicating that they system has either infinitely many solutions or none. When we examine the third row of $rref(A)$ we see a problem.</p>
<p>Remember that there is a linear system behind the matrix and the vertical line in the agumented matrix stands in for the equal sign. The last row states $0=1$. This is not the case, therefore, the system is inconsistent and has no solution.</p>

<p>As you may have guessed this will be geometrically represnted as two lines that never intersect.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Set x as a 5 equaly spaced values between -10 and 10
x = np.linspace(-10,10,5)

# Set y1 and y2 as the equations in the linear system (converted to slope intercept form)
y1 = -1/2 * x + 3/4
y2 = -1/2 * x + 1/3


plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
plt.title('Graph of a linear system with no solution', fontsize=18)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
```

Let's look a three dimensional version of the problem.

$$
\begin{align}
\text{linear system} &=
\begin{vmatrix} 
    x + 4y + z = 0 \\
    4x + 13y + 7z = 0 \\
    7x + 22y + 13z = 1
\end{vmatrix}
\\ 
\\
A &= 
\left[
\begin{array}{ccc|c}
    1 & 4 & 1 & 0 \\
    4 & 13 & 7 & 0 \\
    7 & 22 & 13 & 1
\end{array}
\right]
\\ 
\\
rref(A) &= 
\left[
\begin{array}{ccc|c}
    1 & 0 & 5 & 0 \\
    0 & 1 & -1 & 0 \\
    0 & 0 & 0 & 1
\end{array}
\right]
\end{align}
$$

<p>Again, the coefficient matrix $rank(A) < m$, $(2<3)$ and the last row states $0=1$, and by consequence this linear system has no solution</p> 

<p>Geometrically, this linear system is three planes and there is no point at which the three planes intersect.</p>

```python
# Magic command to allow for interactive plots
%matplotlib notebook

# Import additional libraries for 3d plotting
from mpl_toolkits.mplot3d import Axes3D

# Create a 3x4 matrix with the entries above
A = np.array([[1, 4, 1, 0],[1, 13, 7, 0], [7, 2, 13, 1]])

# Create grid space for 3D plotting
x_vals = np.linspace(-3, 3, 30) 
y_vals = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x_vals, y_vals)

# Create the eqautions for the planes described by the linear system by solving for z
# Note that python uses zero indexing so that the entry A(1,1) is coded as A(0,0)
Z = ((-A[0,0])/A[0,2]) * X + (-A[0,1]/A[0,2]) * Y + (A[0,3]/A[0,2])
Z2 = (-A[1,0]/A[1,2]) * X + (-A[1,1]/A[1,2]) * Y + (A[1,3]/A[1,2])
Z3 = (-A[2,0]/A[2,2]) * X + (-A[2,1]/A[2,2]) * Y + (A[2,3]/A[2,2])

# Create 3D plotter attached to new figure
ax = Axes3D(plt.figure()) 

# Plot the planes
ax.plot_surface(X, Y, Z, alpha=0.3) 
ax.plot_surface(X, Y, Z2, alpha=0.3)
ax.plot_surface(X, Y, Z3, alpha=0.3) 

# Plot settings
ax.view_init(15, 134)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.autoscale(enable=True, axis='both', tight=True)
```

## 3.  Problems with Infinitely Many Solutions

<p>Given the following matrix how do we interpret its solution?</p>

$$
\begin{align}
\text{linear system} &=
\begin{vmatrix} 
    x + 2y + 3z = 1 \\
    3x + 2y + z = 1 \\
    7x + 2y - 3z = 1
\end{vmatrix}
\\
\\
A &=
\left[
\begin{array}{ccc|c}
    1 & 2 & 3 & 1 \\
    3 & 2 & 1 & 1 \\
    7 & 2 & -3 & 1
\end{array}
\right]
\\
\\
rref(A) &=
\left[
\begin{array}{ccc|c}
    1 & 0 & -1 & 0 \\
    0 & 1 & 2 & \frac{1}{2} \\
    0 & 0 & 0 & 0
\end{array}
\right]
\end{align}
$$

<p>As above, let's use the size and rank of the matrix to help us assess the situation.  Again, we notice that in regards to the coefficient matrix $rank(A) < m$, $(2<3)$, but this time we find the last row to give use the result $0=0$. As this is true $rref(A)$ is consistent.</p>
<p>When we try to interpret the result we find a difficulty.  If we read the first, second, and third colums as our $x$, $y$, and $z$ variables from the linear system we arrive a the following conclusions:</p>

$$
\begin{vmatrix} 
    x  &-z &= 0 \\
    \qquad y &+ 2z &= \frac{1}{2} 
\end{vmatrix}
$$

<p>How do we solve this?</p>
<p>As the variable $z$ is shared in both equations it is a <em>free variable</em> and we can set it to an aribrary number we shall call $t$. Therfore, if $z=t$:<p>
    
$$
\begin{matrix}
    \begin{vmatrix} 
        x  &-t &= 0 \\
        \qquad y &+ 2t &= \frac{1}{2} 
    \end{vmatrix}
    &
    \longrightarrow
    &
    \begin{vmatrix} 
        x = t \\
        y = \frac{1}{2} - 2t
    \end{vmatrix}
\end{matrix}
$$
    
<p>Whatever value we give to $t$, we will get a value for $x$ and $y$. As there is an infintie amount of values we can give to $t$, there are an infinite amount of possible answers for $x$ and $y$. This system as infinitely many solutions.</p>

<p>The geometric result is two planes that intersect at a line, the line (stretching indefinitley in either direction) being the respresentation of the infinte solutions available.</p>

```python
# Magic command to allow for interactive plots
%matplotlib notebook

# Libraries needed
import numpy as np
import matplotlib.pyplot as plt
import sympy
from mpl_toolkits.mplot3d import Axes3D

# Array to be plotted
A = np.array([[1, 2, 3, 1], [3, 2, 1, 1], [7, 2, -3, 1]])

# Use sympy to convert the array to reduced row eschelon form (rref)
B = sympy.Matrix(A).rref()

# Convert the rref matrix back to a numpy array
C = np.array(B[0].tolist(), dtype=float)

# Set values for the arbitrary variable t
t = np.linspace(-5, 5, 20)

# Using the rref matrix to determine the equations for the leading variables
x = -C[0,1] * t - C[0,2] * t + C[0,3]
y = -C[1,2] * t + C[1,3]
z = t

# Create grid space for 3D plotting
x_vals = np.linspace(-5, 5, 30) 
y_vals = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x_vals, y_vals)

# Create the eqautions for the planes described by the linear system
Z = ((-A[0,0])/A[0,2]) * X + (-A[0,1]/A[0,2]) * Y + (A[0,3]/A[0,2])
Z2 = (-A[1,0]/A[1,2]) * X + (-A[1,1]/A[1,2]) * Y + (A[1,3]/A[1,2])
Z3 = (-A[2,0]/A[2,2]) * X + (-A[2,1]/A[2,2]) * Y + (A[2,3]/A[2,2])

# Plot panes
ax = Axes3D( plt.figure(figsize=(5,5))) 
ax.plot_surface(X, Y, Z, alpha=0.5) 
ax.plot_surface(X, Y, Z2, alpha=0.5) 
ax.plot_surface(X, Y, Z3, alpha=0.5) 

# Plot line of infinite solutions
ax.plot(x,y,z, "--")

# Plot settings
ax.view_init(15, 134)
ax.set_title('Plot of a linear system', fontsize=15)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.autoscale(enable=True, axis='both', tight=True)

plt.show()
```
